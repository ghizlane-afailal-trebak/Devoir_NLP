## Extraction du texte depuis le PDF

```python
import PyPDF2

def extract_text_from_pdf(pdf_path):
    """Extrait tout le texte d'un fichier PDF page par page."""
    text = ""
    with open(pdf_path, "rb") as f:
        reader = PyPDF2.PdfReader(f)
        for page in reader.pages:
            text += page.extract_text() + "\n"
    return text

# --- Charger le PDF arabe ---
PDF_PATH = "exemple_text_climate.pdf"  
raw_text = extract_text_from_pdf(PDF_PATH)

print("=== TEXTE BRUT EXTRAIT DU PDF ===")
print(raw_text[:800])
```

#  Version 1 : Pipeline NLP — Analyse Approfondie d’un Texte Climatique en Arabe
1. Tokenisation (التقطيع)

Le texte arabe brut est segmenté en unités lexicales (tokens) après une phase de normalisation orthographique.

Observation :
Des variations comme "الأرض" et "الارض" sont unifiées grâce à la normalisation des différentes formes de l’alif. De plus, les diacritiques (tashkīl), souvent absents ou inconsistants, sont supprimés afin de garantir une représentation homogène des mots.
Objectif :
Réduire les variations superficielles de l’écriture arabe afin que des mots identiques sur le plan sémantique soient traités comme une seule entité. Cela améliore la cohérence des étapes suivantes du traitement.
2. Suppression des Stopwords (حذف الكلمات الوظيفية)

Les mots fonctionnels (stopwords), qui n’apportent pas de valeur sémantique significative, sont éliminés.

Observation :
Des termes tels que ['إلى', 'في', 'على', 'من', 'مما'] sont supprimés, car ils jouent un rôle grammatical mais n’apportent pas d’information thématique.
Objectif :
Conserver uniquement les mots porteurs de sens afin de mieux capturer le contenu du texte.
Dans un contexte climatique, les mots importants deviennent :
['ارتفاع', 'درجات', 'الحرارة', 'ذوبان', 'الجليد', ...]

 Cela permet de concentrer l’analyse sur le contenu informatif réel.

3. Racinisation avec ISRIStemmer (استخراج الجذور)

Le stemming permet de réduire les mots à leur forme racine, facilitant ainsi la généralisation linguistique.

Transformations observées :
ذوبان → ذوب
المحيطات → محيط
البحار → بحر
يسبب → سبب
Objectif :
Regrouper différentes formes d’un même mot sous une racine commune.
Ainsi, un modèle entraîné sur "بحر" pourra reconnaître automatiquement "البحار", "بحري" ou "بحور".

 Remarque importante :
L’ISRIStemmer est un outil de stemming et non de lemmatisation. Il peut produire des racines non lexicalisées (non présentes dans le dictionnaire), mais reste efficace pour des méthodes statistiques comme TF-IDF.

4. Vectorisation TF-IDF (الترميز الرياضي)

Le texte est transformé en vecteur numérique à l’aide de la méthode TF-IDF (Term Frequency – Inverse Document Frequency).

Cette méthode repose sur deux composantes :

TF (Term Frequency) : fréquence d’un mot dans le document
IDF (Inverse Document Frequency) : rareté du mot dans le corpus
   Analyse des scores TF-IDF (texte climatique)
Score élevé :
Des mots comme 'جليد' (glace) ou 'منسوب' (niveau) obtiennent un score élevé car ils sont spécifiques au domaine climatique et rares dans les autres documents.
Score moyen :
Des mots comme 'حرارة' ou 'بيئة' sont importants mais apparaissent dans plusieurs documents, ce qui diminue leur poids.
Score nul (0.00) :
Le mot n’apparaît pas dans le document analysé, même s’il existe dans le vocabulaire global.

```python
import re
import pandas as pd
from nltk.corpus import stopwords
from nltk.stem.isri import ISRIStemmer
from sklearn.feature_extraction.text import TfidfVectorizer

# Outils
arabic_stopwords = set(stopwords.words('arabic'))
stemmer = ISRIStemmer()

def normalize_arabic(text):
    """Normalise le texte arabe avant tokenisation."""
    text = re.sub(r'[\u0610-\u061A\u064B-\u065F]', '', text)  # Diacritiques
    text = re.sub(r'[إأآا]', 'ا', text)   # Alifs
    text = re.sub(r'ى', 'ي', text)         # Ya' final
    text = re.sub(r'ة', 'ه', text)         # Ta' marbuta
    text = re.sub(r'[^\u0600-\u06FF\s]', '', text)  # Non-arabe
    return text.strip()

# ===========================================================
# STEP 1 : Normalisation & Tokenisation
# ===========================================================
texte_norm = normalize_arabic(phrase_test)
tokens_bruts = texte_norm.split()

print("=" * 60)
print(f"TEXTE ORIGINAL  : {phrase_test}")
print(f"TEXTE NORMALISÉ : {texte_norm}")
print("=" * 60)
print(f"\n1. Tokenisation ({len(tokens_bruts)} tokens) :")
for i, t in enumerate(tokens_bruts):
    print(f"   [{i:02d}] '{t}'")

# ===========================================================
# STEP 2 : Retrait des Stopwords
# ===========================================================
tokens_filtres = [t for t in tokens_bruts if t not in arabic_stopwords]
supprimes = [t for t in tokens_bruts if t in arabic_stopwords]

print(f"\n2. Après retrait des Stopwords :")
print(f"   Conservés ({len(tokens_filtres)}) : {tokens_filtres}")
print(f"   Supprimés ({len(supprimes)}) : {list(set(supprimes))}")
print(f"   Taux de réduction : {len(supprimes)/len(tokens_bruts)*100:.1f}% de bruit éliminé")

# ===========================================================
# STEP 3 : Racinisation — ISRIStemmer
# ===========================================================
racines = [stemmer.stem(t) for t in tokens_filtres]

print(f"\n3. Racinisation (ISRIStemmer) :")
print(f"   {'Mot original':<30} → {'Racine extraite':<20}")
print(f"   {'-'*52}")
for mot, racine in zip(tokens_filtres, racines):
    marker = " ✓" if mot != racine else "  (inchangé)"
    print(f"   {mot:<30} → {racine:<20}{marker}")

# ===========================================================
# STEP 4 : TF-IDF
# ===========================================================
corpus = [
    " ".join(racines),              # Notre document — changement climatique
    "غاب حرق وقود مصنع تلوث هواء", # Doc 2 — pollution industrielle
    "طاق شمس ريح بيئ استدام تجدد"  # Doc 3 — énergies renouvelables
]

vec = TfidfVectorizer()
mat = vec.fit_transform(corpus)

df = pd.DataFrame(
    mat.toarray(),
    columns=vec.get_feature_names_out(),
    index=["Changement climatique", "Pollution industrielle", "Énergies renouvelables"]
)

print(f"\n4. Représentation TF-IDF — Changement climatique (scores décroissants) :")
scores = df.iloc[0].sort_values(ascending=False)

print("\n--- Tableau comparatif (tous documents) ---")
print(df.to_string())
```

##  VERSION 2 — Fonction Autonome (Approche Code-First)

Dans cette version, nous implémentons une fonction complète qui encapsule toute la pipeline NLP.

Contrairement à l’approche étape par étape précédente, cette version adopte une approche **"Code-First"** :
toutes les transformations (tokenisation, nettoyage, stemming, vectorisation) sont regroupées dans une seule fonction.

###  Objectifs :

- Automatiser le traitement d’un texte arabe
- Rendre la pipeline réutilisable sur n’importe quel texte
- Simplifier l’appel des différentes étapes NLP
- Faciliter l’intégration dans un système réel (application, API, etc.)

###  Fonctionnalités :

La fonction effectue les étapes suivantes :

1. Normalisation du texte arabe  
2. Tokenisation  
3. Suppression des stopwords  
4. Stemming avec ISRIStemmer  
5. Transformation en vecteur TF-IDF  
6. Prédiction avec le modèle entraîné  



```python
import re
import pandas as pd
from nltk.corpus import stopwords
from nltk.stem.isri import ISRIStemmer
from sklearn.feature_extraction.text import TfidfVectorizer

# Chargement des outils arabes
arabic_stopwords = set(stopwords.words('arabic'))
stemmer = ISRIStemmer()

def normalize_arabic(text):
    text = re.sub(r'[\u0610-\u061A\u064B-\u065F]', '', text)
    text = re.sub(r'[إأآا]', 'ا', text)
    text = re.sub(r'ى', 'ي', text)
    text = re.sub(r'ة', 'ه', text)
    text = re.sub(r'[^\u0600-\u06FF\s]', '', text)
    return text.strip()

def pipeline_nlp_ar(text):
    """
    Applique et trace un pipeline NLP complet sur un texte arabe :
    1. Normalisation & Tokenisation
    2. Retrait des Stopwords
    3. Racinisation (ISRIStemmer)
    Retourne le texte raciné, prêt pour TF-IDF.
    """
    print(f"--- TEXTE BRUT ---\n'{text}'\n")

    # 1. NORMALISATION & TOKENISATION
    texte_norm = normalize_arabic(text)
    tokens_bruts = texte_norm.split()
    print(f"1. Tokenisation (normalisé) :\n   {tokens_bruts}\n")

    # 2. FILTRAGE — Stopwords arabes
    tokens_filtres = [t for t in tokens_bruts if t not in arabic_stopwords]
    supprimes = [t for t in tokens_bruts if t in arabic_stopwords]
    print(f"2. Après retrait des Stopwords :")
    print(f"   {tokens_filtres}")
    print(f"   (Mots supprimés : {list(set(supprimes))})\n")

    # 3. RACINISATION — ISRIStemmer
    # Ramène chaque mot à sa racine trilitère arabe
    racines = [stemmer.stem(t) for t in tokens_filtres]
    print(f"3. Racinisation (Racines trilitères) :")
    print(f"   {racines}\n")

    return " ".join(racines)

# --- EXEMPLE DÉTAILLÉ avec la phrase extraite du PDF ---
phrase_test = "ارتفاع درجات الحرارة على سطح الأرض يؤدي إلى ذوبان الجليد في القطبين مما يسبب ارتفاع منسوب مياه البحار والمحيطات"
texte_propre = pipeline_nlp_ar(phrase_test)

# --- 4. ZOOM SUR TF-IDF ---
# Mini-corpus autour du thème environnemental pour comparaison
corpus = [
    texte_propre,
    "غاب حرق وقود مصنع تلوث هواء",
    "طاق شمس ريح بيئ استدام تجدد"
]

vectorizer = TfidfVectorizer()
tfidf_matrix = vectorizer.fit_transform(corpus)

# Affichage des scores pour le premier document
print("4. Représentation Numérique (TF-IDF) pour le document 1 :")
df_tfidf = pd.DataFrame(
    tfidf_matrix.toarray(),
    columns=vectorizer.get_feature_names_out()
)
print(df_tfidf.iloc[0].sort_values(ascending=False))
```
